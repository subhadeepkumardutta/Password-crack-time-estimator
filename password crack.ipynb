```python
import tkinter as tk
import math

# Function to estimate crack time
def estimate_time():
    password = entry.get()
    length = len(password)

    # Character set size estimation
    charset = 0
    if any(c.islower() for c in password):
        charset += 26
    if any(c.isupper() for c in password):
        charset += 26
    if any(c.isdigit() for c in password):
        charset += 10
    if any(not c.isalnum() for c in password):
        charset += 32

    if charset == 0:
        result_label.config(text="Please enter a password")
        return

    # Total combinations
    combinations = charset ** length

    # Assume attacker tries 1 billion guesses per second
    guesses_per_second = 1_000_000_000

    seconds = combinations / guesses_per_second

    # Convert time
    minutes = seconds / 60
    hours = minutes / 60
    days = hours / 24
    years = days / 365

    result = f"""
Estimated Crack Time:
Seconds: {seconds:.2e}
Minutes: {minutes:.2e}
Hours: {hours:.2e}
Days: {days:.2e}
Years: {years:.2e}
"""

    result_label.config(text=result)

# GUI Window
root = tk.Tk()
root.title("Password Crack Time Estimator")
root.geometry("400x300")

title = tk.Label(root, text="Password Crack Time Estimator", font=("Arial", 14))
title.pack(pady=10)

label = tk.Label(root, text="Enter Password:")
label.pack()

entry = tk.Entry(root, width=30, show="*")
entry.pack(pady=5)

button = tk.Button(root, text="Estimate Crack Time", command=estimate_time)
button.pack(pady=10)

result_label = tk.Label(root, text="", justify="left")
result_label.pack(pady=10)

root.mainloop()

```
